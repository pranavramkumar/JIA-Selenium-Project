```python
# populate
```

```python
for ind, row in df.iterrows():
    #execboard
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu'+df.iloc[ind,0])
    time.sleep(2)
    elem = browser.find_element_by_xpath('//*[@id="sidebar-node"]/div/div[2]/ul/li[1]')
    elem.click()
    elems = browser.find_elements_by_xpath("//div[@class='paragraphs-dropbutton-wrapper']")
    elems[1].click()
    time.sleep(3)
    framedriver = browser.find_element_by_id('cke_1_contents').find_element_by_tag_name('iframe')
    browser.switch_to.frame(framedriver)
    cke = browser.find_element_by_tag_name('body')
    cke.clear()
    cke.send_keys(df.iloc[ind,2])
    browser.switch_to.default_content()
    time.sleep(2)
    elem = browser.find_element_by_id('edit-submit')
    elem.click()
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu')
    #sectioneditors
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu'+df.iloc[ind,0])
    time.sleep(2)
    elem = browser.find_element_by_xpath('//*[@id="sidebar-node"]/div/div[2]/ul/li[1]')
    elem.click()
    elems = browser.find_elements_by_xpath("//div[@class='paragraphs-dropbutton-wrapper']")
    elems[2].click()
    time.sleep(3)
    framedriver = browser.find_element_by_id('cke_1_contents').find_element_by_tag_name('iframe')
    browser.switch_to.frame(framedriver)
    cke = browser.find_element_by_tag_name('body')
    cke.clear()
    cke.send_keys(df.iloc[ind,3])
    browser.switch_to.default_content()
    time.sleep(2)
    elem = browser.find_element_by_id('edit-submit')
    elem.click()
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu')
    #editorialboard
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu'+df.iloc[ind,0])
    time.sleep(2)
    elem = browser.find_element_by_xpath('//*[@id="sidebar-node"]/div/div[2]/ul/li[1]')
    elem.click()
    elems = browser.find_elements_by_xpath("//div[@class='paragraphs-dropbutton-wrapper']")
    elems[3].click()
    time.sleep(3)
    framedriver = browser.find_element_by_id('cke_1_contents').find_element_by_tag_name('iframe')
    browser.switch_to.frame(framedriver)
    cke = browser.find_element_by_tag_name('body')
    cke.clear()
    cke.send_keys(df.iloc[ind,4])
    browser.switch_to.default_content()
    time.sleep(2)
    elem = browser.find_element_by_id('edit-submit')
    elem.click()
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu')
```
