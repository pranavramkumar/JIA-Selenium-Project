```python
for ind, row in df.iterrows():
    #arguments
    browser.get('https://jia.site.drupaldisttest.cc.columbia.edu'+df.iloc[ind,1])
    elem = browser.find_element_by_link_text('Edit')
    elem.click()
    time.sleep(4)
    #open TOC accordion
    elem = browser.find_element_by_xpath("//div[contains(@id,'edit-field-cu-content-0')]//div[@class='paragraphs-dropbutton-wrapper']")
    elem.click()
    time.sleep(4)
    #open arguments
    elem = browser.find_element_by_xpath("//div[contains(@id,'edit-field-cu-content-0-subform-field-cu-accordion-content-2')]//div[@class='paragraphs-dropbutton-wrapper']")
    elem.click()
    time.sleep(4)
    elem = browser.find_element_by_xpath("//div[contains(@id,'edit-field-cu-content-0-subform-field-cu-accordion-content-2-subform-field-cu-accordion-item-content-0')]//div[@class='paragraphs-dropbutton-wrapper']")
    elem.click()
    time.sleep(4)
    elems = browser.find_elements_by_xpath("//div[@class='cke_inner cke_reset']//iframe[contains(@class,'cke_wysiwyg_frame')]")
    framedriver = elems[1]
    browser.switch_to.frame(framedriver)
    time.sleep(4)
    cke = browser.find_element_by_tag_name('body')
    cke.clear()
    if str(df.iloc[ind,4]) != 'nan':
        cke.send_keys(df.iloc[ind,4])
    browser.switch_to.default_content()
    time.sleep(4)
    #submit
    elem = browser.find_element_by_id('edit-submit')
    elem.click()
```
