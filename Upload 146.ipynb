```python
for ind, row in df.iterrows():
    # open and click edit
    browser.get(df.iloc[ind,2])
    elem = browser.find_element_by_link_text('Edit')
    elem.click()
    time.sleep(3)
    
   # clear summary
    elems = browser.find_elements_by_xpath("//div[@id='cke_2_contents']//iframe[contains(@class,'cke_wysiwyg_frame')]")
    framedriver = elems[0]
    browser.switch_to.frame(framedriver)
    time.sleep(3)
    cke = browser.find_element_by_tag_name('body')
    time.sleep(3)
    cke.clear()
    browser.switch_to.default_content()
    
    # submit
    elem = browser.find_element_by_id('edit-submit')
    elem.click()
```
